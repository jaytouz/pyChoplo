```python
import matplotlib
from matplotlib.ticker import FormatStrFormatter

font = {'family' : 'normal',
        'weight': 'normal',
    'size'   : 16}
fig = plt.figure()
ytick = list("Joueur {}".format(i) for i in range(1,11))
xtick = list("Niveau {}".format(i) for i in range(1, 4))
g = sns.heatmap(v_missing_df, vmin=0, vmax=4, cmap='RdYlGn_r', linewidths=1, linecolor='k', annot=True)
plt.title('Mouvements retirés sur 30 selon le joueur et le niveau')
g.set_yticklabels(ytick)
g.set_xticklabels(xtick)
fig.set_size_inches(9.5, 6.5)
plt.savefig('Vitesse mouvement retire.png', dpi=1000)
```

```python
SMALL_SIZE = 16
MEDIUM_SIZE = 18
BIGGER_SIZE = 20

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE) 



fig, axes = plt.subplots(nrows=1,ncols=2, sharey=True, sharex = True)
fig.set_size_inches(18.5,10)
ytick = list("Joueur {}".format(i) for i in range(1,11))
xtick = list("Niveau {}".format(i) for i in range(1, 4))

g1 = sns.heatmap(v_missing_df, vmin=0, vmax=4, cmap='RdYlGn_r', linewidths=1, cbar=False, linecolor='k', yticklabels = ytick, xticklabels=xtick, annot=True, ax=axes[0], )
g2 = sns.heatmap(a_missing_df,vmin=0, vmax=4, cmap='RdYlGn_r', linewidths=1, linecolor='k', yticklabels = ytick, xticklabels=xtick, annot=True, ax=axes[1])
g1.set_title('Vitesse', fontsize=BIGGER_SIZE, weight='bold')
g2.set_title('Amplitude', fontsize=BIGGER_SIZE, weight = 'bold')

cbar_ax = fig.axes[-1]
cbar_solids = cbar_ax.set_frame_on(True)

plt.suptitle('Mouvements retirés sur 30 selon le joueur et le niveau', fontsize=26, y=1.05)
plt.tight_layout()

plt.savefig('missing_values.png', dpi=1000)
```
